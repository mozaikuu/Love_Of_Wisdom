# Starting with importing the data

```python
import pandas as pd

# Load the CSV file into a DataFrame
dfs = pd.read_csv('./Datasets/Fake_Shocking_small(5000)_dataset.csv')
dfm = pd.read_csv('./Datasets/Fake_Shocking_medium(50000)_dataset.csv')
dfl = pd.read_csv('./Datasets/Fake_Shocking_large(500000)_dataset.csv')
```

# explore the data

```python
df_list = [dfs, dfm, dfl]
def Print_df():
    for df in df_list:
        print(df.info())
        print(df.describe())
        print(df.shape)
        print(df.columns)
        print(df.dtypes)
        print(df.isnull().sum())
        print(df)
        print(f" -----------------------------------END OF DATAFRAME-----------------------------------")
        

Print_df()
```

```python
for df in df_list_numeric:
    sns.heatmap(df.corr())
    plt.show()
```

```python
for df in df_list_numeric:
    correlation_matrix = df.corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Matrix')
    plt.show()
    print(f" -----------------------------------END OF DATAFRAME-----------------------------------")
```
